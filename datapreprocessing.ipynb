```python
def flatten_h5_file(file_path):
    with h5py.File(file_path, 'r') as file:
        all_data = []

        def extract_datasets(name, obj):
            if isinstance(obj, h5py.Dataset):
                try:
                    # Convert dataset to DataFrame
                    if obj.shape == ():  # Handle scalar datasets
                        data = pd.DataFrame([obj[()]], columns=[name])
                    elif len(obj.shape) == 1:  # Handle 1D array datasets
                        data = pd.DataFrame(obj[:], columns=[name])
                    else:  # Handle multi-dimensional datasets
                        data = pd.DataFrame(obj[:,:].flatten(), columns=[name])

                    return data
                except Exception as e:
                    print(f"Error processing {name} in {file_path}: {e}")
                    return None

        # Extract and concatenate all datasets from the file
        datasets = [extract_datasets(name, obj) for name, obj in file.items() if isinstance(obj, h5py.Dataset)]
        datasets = [d for d in datasets if d is not None]  # Remove None datasets

        if datasets:
            file_df = pd.concat(datasets, axis=1)
            file_df['source_file'] = file_path.split('/')[-1]  # Add a column to identify the source file
            return file_df
        else:
            print(f"No valid datasets found in {file_path}")
            return pd.DataFrame()

# Updating the loop for reading files
all_data = pd.DataFrame()
for file_path in file_paths:
    file_data = flatten_h5_file(file_path)
    if not file_data.empty:
        all_data = pd.concat([all_data, file_data], ignore_index=True)
```

```python
import h5py

# Specify the path to the HDF5 file
h5_file_path = 'TRAAAMO128F1481E7Fnnnnn.h5'

# Function to recursively print the structure of the HDF5 file
def print_h5_structure(group, level=0):
    # Indentation for clarity
    indent = "  " * level

    # Iterate through the items (datasets and subgroups) in the group
    for item_name in group:
        item = group[item_name]
        if isinstance(item, h5py.Dataset):
            print(f"{indent}Dataset: {item_name}")
        elif isinstance(item, h5py.Group):
            print(f"{indent}Group: {item_name}")
            # Recursively print the structure of subgroups
            print_h5_structure(item, level + 1)

# Open the HDF5 file for reading
with h5py.File(h5_file_path, 'r') as h5_file:
    print("Structure of the HDF5 file:")
    print_h5_structure(h5_file)

# This code will print the structure of the HDF5 file, including groups and datasets.
```

```python
import h5py
import pandas as pd
import os

def save_dataset_as_csv(file, group_name, dataset_name, output_folder):
    """Save a given dataset as a CSV file."""
    dataset_data = file[group_name][dataset_name][:]
    df = pd.DataFrame(dataset_data)
    csv_file_name = f"{output_folder}/{group_name}_{dataset_name}.csv"
    df.to_csv(csv_file_name, index=False)
    print(f"Saved {csv_file_name}")

def process_hdf5_file(hdf5_file_path, output_folder):
    """Process each dataset in each group of the HDF5 file and save them as CSV files."""
    with h5py.File(hdf5_file_path, 'r') as file:
        # Ensure the output folder exists
        if not os.path.exists(output_folder):
            os.makedirs(output_folder)

        # Iterate through each group and dataset in the HDF5 file
        for group_name in file.keys():
            print(f"Processing group: {group_name}")
            for dataset_name in file[group_name].keys():
                print(f"  Processing dataset: {dataset_name}")
                save_dataset_as_csv(file, group_name, dataset_name, output_folder)

# Usage
hdf5_file_path = 'TRBIDFH128F1464397.h5'  # Replace with the actual path to your HDF5 file
output_folder = 'Untitled Folder'      # Folder where CSV files will be saved
process_hdf5_file(hdf5_file_path, output_folder)
```

```python
import os

def find_hdf5_files(directory):
    """Find all HDF5 (.h5) files in the specified directory and its subdirectories."""
    hdf5_files = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.h5'):
                hdf5_files.append(os.path.join(root, file))
    return hdf5_files

# Specify the directory where your HDF5 files are located
directory = '/Users/prithvik/Downloads/MillionSongSubset'  # Replace with the path to your directory

# Find all HDF5 files in the directory and its subdirectories
hdf5_file_paths = find_hdf5_files(directory)

print(hdf5_file_paths)
```
